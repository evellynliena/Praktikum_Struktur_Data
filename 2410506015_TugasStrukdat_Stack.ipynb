```python
class TowerOfHanoi:
    def __init__(self, num_disks, max_steps):
        self.towers = {
            'A': list(range(num_disks, 0, -1)),
            'B': [],
            'C': []
        }
        self.num_disks = num_disks
        self.max_steps = max_steps
        self.steps_taken = 0

    def move_disk(self, from_tower, to_tower):
        if self.steps_taken >= self.max_steps:
            print("Batas langkah maksimal tercapai!")
            return

        if not self.towers[from_tower]:
            print(f"Tidak ada cakram di tiang {from_tower} untuk dipindahkan!")
            return

        if self.towers[to_tower] and self.towers[to_tower][-1] < self.towers[from_tower][-1]:
            print("Tidak bisa menaruh cakram besar di atas cakram kecil!")
            return

        disk = self.towers[from_tower].pop()
        self.towers[to_tower].append(disk)
        self.steps_taken += 1
        print(f"Memindahkan cakram {disk} dari {from_tower} ke {to_tower}")

    def display_towers(self):
        print("\nKondisi Tiang:")
        for key, value in self.towers.items():
            print(f"{key}: {value}")
        print(f"Langkah: {self.steps_taken}/{self.max_steps}")

    def is_solved(self):
        return len(self.towers['B']) == self.num_disks


num_disks = 3
max_steps = 15

game = TowerOfHanoi(num_disks, max_steps)


while not game.is_solved():
    game.display_towers()
    print("\nðŸ”¹ Pilih langkah: ")
    from_tower = input("Pindahkan dari tiang (A/B/C): ").upper()
    to_tower = input("Ke tiang (A/B/C): ").upper()

    if from_tower in "ABC" and to_tower in "ABC" and from_tower != to_tower:
        game.move_disk(from_tower, to_tower)
    else:
        print("Masukkan tidak valid! Harap pilih antara A, B, atau C.")

    if game.steps_taken >= max_steps:
        print("\nBatas langkah tercapai! Game Over.")
        break

if game.is_solved():
    print("\nSelamat! Anda menyelesaikan Tower of Hanoi!")
```
