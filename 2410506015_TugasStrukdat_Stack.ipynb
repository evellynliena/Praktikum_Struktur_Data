```python
class WarehouseManager:
    def __init__(self, max_capacity):
        self.stack = []
        self.max_capacity = max_capacity

    def push(self, item):
        if len(self.stack) < self.max_capacity:
            self.stack.append(item)
            print(f"Barang '{item}' ditambahkan ke gudang.")
        else:
            print("Gudang penuh! Tidak bisa menambahkan barang lagi.")

    def pop(self):
        if self.stack:
            removed_item = self.stack.pop()
            print(f"Barang '{removed_item}' telah diambil dari gudang.")
        else:
            print("Gudang kosong! Tidak ada barang untuk diambil.")

    def show_items(self):
        if self.stack:
            print("\nDaftar Barang dalam Gudang:")
            for i, item in enumerate(reversed(self.stack), 1):
                print(f"{i}. {item}")
        else:
            print("Gudang kosong!")


npm = int(input("Masukkan 2 nomor NPM akhir: "))
warehouse = WarehouseManager(npm)

while True:
    print("\nMenu Gudang:")
    print("1. Tambah Barang")
    print("2. Ambil Barang")
    print("3. Lihat Isi Gudang")
    print("4. Keluar")

    pilihan = input("Pilih opsi (1-4): ")

    if pilihan == "1":
        barang = input("Masukkan nama barang: ")
        warehouse.push(barang)
    elif pilihan == "2":
        warehouse.pop()
    elif pilihan == "3":
        warehouse.show_items()
    elif pilihan == "4":
        print("Terima kasih!")
        break
    else:
        print("Pilihan tidak valid! Silakan pilih opsi 1-4.")
```
